```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
cars_data = pd.read_csv('electric_vehicles_spec_2025.csv.csv')
```

```python
cars_data.head()
```

```python
cars_data.drop(
    columns=[
        'battery_type', 'number_of_cells', 'torque_nm', 'towing_capacity_kg',
        'drivetrain', 'segment', 'length_mm', 'width_mm', 'height_mm',
        'source_url', 'fast_charge_port', 'cargo_volume_l'
    ],
    inplace=True,
    errors='ignore'   # <-- This avoids KeyError
)
```

```python
#Data Analysis
```

```python
for col in cars_data.columns:
    print('Unique values of' + col)
    print(cars_data[col].unique())
    print("==========================")
```

```python
cars_data.drop(
    columns=[
        'model'
    ],
    inplace=True,
    errors='ignore'   # <-- This avoids KeyError
)
```

```python
for col in cars_data.columns:
    print('Unique values of ' + col)
    print(cars_data[col].unique())
    print("==========================")
```

```python
# Get unique brand names
brands = cars_data['brand'].unique()

# Create a dictionary that assigns each brand a unique number
brand_mapping = {brand: idx+1 for idx, brand in enumerate(brands)}

# Replace brand names with numbers
cars_data['brand'] = cars_data['brand'].map(brand_mapping)
```

```python
# Get unique body type values
body_types = cars_data['car_body_type'].unique()

# Create mapping dictionary
body_mapping = {body: idx+1 for idx, body in enumerate(body_types)}

# Replace body type names with numbers
cars_data['car_body_type'] = cars_data['car_body_type'].map(body_mapping)

# Show mapping
body_mapping
```

```python
for col in cars_data.columns:
    print('Unique values of ' + col)
    print(cars_data[col].unique())
    print("==========================")
```

```python
cars_data
```

```python
input_data = cars_data.drop(columns=['efficiency_wh_per_km'])
output_data = cars_data['efficiency_wh_per_km']
```

```python
x_train, x_test, y_train, y_test = train_test_split(input_data, output_data, test_size=0.2)
```

```python
#Model Creation
```

```python
model = LinearRegression()
```

```python
#Train Model
```

```python
model.fit(x_train, y_train)
```

```python
predict = model.predict(x_test)
```
